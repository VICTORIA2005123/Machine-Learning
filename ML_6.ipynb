```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
col_names = ["sepallength", "sepalwidth", "petallength", "petalwidth", "class"]
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
dataset = pd.read_csv(url,names=col_names)
```

```python
dataset
```

```python
dataset_knn = dataset.values[:,0:4]
labels = dataset.values[:,-1]
```

```python
dataset_knn
```

```python
# Define the function to find the distance
import math
def distance(x1,x2):
  distances = 0.0
  for i in range(len(x1)):
    distances += (x1[i]-x2[i])**2
  return (math.sqrt(distances))
```

```python
testing_data = dataset.values[5,0:4]
```

```python
distance_cumm = []
for i in range(len(dataset_knn)):
  distance_cumm.append([distance(dataset_knn[i],testing_data),labels[i]])
```

```python
distance_cumm
```

```python
distance_cumm.sort()
```

```python
distance_cumm
```

```python
k=8
labels_topk = []
for i in range(1,k+1):
  labels_topk.append(distance_cumm[i][1])
```

```python
labels_topk
```

```python
from collections import Counter
count = Counter(labels_topk)
```

```python
count
```

```python
print("Final predicted label:", count.most_common(1)[0][0])
```
