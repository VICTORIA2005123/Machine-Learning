```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colnames = ['Area', 'Rooms', 'Price']
dataset = pd.read_csv("https://raw.githubusercontent.com/nishithkotak/machine-learning/refs/heads/master/ex1data2.txt", names=colnames)
```

```python
X = dataset.iloc[:,0:2].values
Y = dataset.iloc[:,2].values
```

```python
def euclidean_distance(x1,x2):
  return np.sqrt(np.sum((x1-x2)**2))
```

```python
def knn_fit(X,y):
  return X, y
```

```python
def knn_predict(X_train, y_train, X_test, k=3):
  predictions = []
  for x in X_test:
    distances = [euclidean_distance(x,x_train) for x_train in  X_train]
    k_indices = np.argsort(distances)[:k]
    k_neighbor_vals = y_train[k_indices]
    prediction = np.mean(k_neighbor_vals)
    predictions.append(prediction)
  return np.array(predictions)
```

```python
X_train,y_train = knn_fit(X,Y)
X_test = np.array([[3000,4]]) #The area of the house, number of rooms
predicted_price = knn_predict(X_train,y_train,X_test,k=5)
print("Predicted price:", predicted_price[0])
```
